```python
import numpy as np
```

```python
normal_dist = np.random.normal(loc=50,scale=3,size=100)
normal_dist
```

```python
o = np.array([1000,1500,2000])
o_data=np.concatenate([normal_dist,o])
o_data
```

```python
nm=np.mean(normal_dist)
om=np.mean(o_data)
print(nm)
print(om)
```

```python
import pandas as pd
```

```python
pd.Series({"Normal_Data":normal_dist})
```

```python
s=pd.DataFrame({"Normal_Data":normal_dist})
s
```

```python
s1=pd.DataFrame({"Outlier_Data":o_data})
s1
```

```python
s.kurt()
```

```python
s1.kurt()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.histplot(normal_dist,kde=True)
plt.title("Normal Data")
plt.subplot(1,2,2)
sns.histplot(o_data,kde=True)
plt.title("Outlier Data")
plt.show()
```
