```python
p=d.isnull().mean()
p
```

```python
x=p.index
y=p.values
```

```python
plt.bar(x,y)
plt.title("Null Values")
plt.xlabel("Column name")
plt.ylabel("Null Percentage")
plt.show()
```

```python
d.isnull().mean().plot(kind="bar")
plt.show()
```
