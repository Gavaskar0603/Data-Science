```python
import pandas as pd
```

```python
d=pd.read_csv(r"C:\Users\gavas\Downloads\StudentsPerformance.csv")
d
```

```python
d.columns
```

```python
import  matplotlib.pyplot as plt
```

```python
d.groupby("gender")[['math score', 'reading score',
       'writing score']].mean().plot(kind="bar")
plt.title("Score Difference in gender")
plt.xlabel("Gender")
plt.ylabel("Scores")
plt.show()
```

```python
choco=["5Star","Diary_Milk","Kit-Kat","Others"]
sales=[1500,2500,2000,1000]
plt.bar(choco,sales)
plt.xlabel("Choclates")
plt.ylabel("Sales")
# plt.color_sequences()
plt.show()
```

```python
plt.hist(d["math score"],bins=20)
plt.title("Math Score Distribution")
plt.xlabel("Score")
plt.ylabel("frequency")
plt.show()
```

```python
plt.scatter(d["math score"],d["reading score"])
plt.title("Score plot")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
avg_s=d[['math score',"reading score","writing score"]].mean()
avg_s
```

```python
x=avg_s.index
y=avg_s.values
y
```

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("Cato Scoring")
plt.ylabel("Scoring")
# plt.grid()
plt.show()
```

```python
d["gender"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.show()
```

```python
#box plot
plt.boxplot([d["math score"],d["reading score"],d["writing score"]])
plt.title("maths vs reading")
plt.xlabel("math score")
plt.ylabel("reading score")
plt.show()
```
