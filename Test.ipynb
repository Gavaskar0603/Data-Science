```python
#line chart
import matplotlib.pyplot as plt
x=["Alice","Bob","Charle","Denver","Eva"]
y=[300,290,356,499,588]
plt.plot(x,y)
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.show()
```
